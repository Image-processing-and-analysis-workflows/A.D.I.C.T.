# 1. 2. Celltool analysis bash pipeline
## Rifampicin (RIF) screening 
### run both bash scripts in the terminal within the respective RIF or CM folder where the binary images are located

```python
#-----extract contours from smoothed png----
celltool extract_contours --min-area=50 --scale=0.100 --units=microns --resample-points=100 --smoothing-factor=0.01 --destination=Contours Binary/*.png

#----plot contours-----
celltool plot_contours --output-file="RIF_nc_Extracted Contours.svg" Contours/*nc*.contour  # _nc_contours
celltool plot_contours --output-file="RIF_after_30_minutes Extracted Contours.svg" Contours/*30min*.contour  # _after_30min_ contours
celltool plot_contours --output-file="RIF_after_60_minutes Extracted Contours.svg" Contours/*60min*.contour  # _after_60min_ contours

#-----align contours-----
celltool align_contours --allow-reflection --destination=Aligned Contours/*nc*.contour
celltool align_contours --allow-reflection --destination=Aligned Contours/*30min*.contour
celltool align_contours --allow-reflection --destination=Aligned Contours/*60min*.contour

#-----plot aligned contours-----
celltool plot_contours --color-by=points --output-file="RIF_nc Aligned Contours.svg" Aligned/*nc*.contour
celltool plot_contours --color-by=points --output-file="RIF_after 30min Aligned Contours.svg" Aligned/*30min*.contour
celltool plot_contours --color-by=points --output-file="RIF_after 60min Aligned Contours.svg" Aligned/*60min*.contour

#----pca---
celltool shape_model --variance-explained=0.97 --output-prefix="RIF_nc model" Aligned/*nc*.contour
celltool shape_model --variance-explained=0.96 --output-prefix="RIF_30min model" Aligned/*30min*.contour
celltool shape_model --variance-explained=0.99 --output-prefix="RIF_60min model" Aligned/*60min*.contour

#----plot pca----
celltool plot_model "RIF_nc model.contour"
celltool plot_model "RIF_30min model.contour"
celltool plot_model "RIF_60min model.contour"

#----plot pca distribution----
celltool plot_distribution --x-column=2 --y-column=3 --output-file="RIF_time course normalized positions.svg" "RIF_nc model-normalized-positions.csv" "RIF_30min model-normalized-positions.csv" "RIF_60min model-normalized-positions.csv" Aligned/*.contour

#---- plot contours ----
celltool plot_contours --color-by=points --label-points --output-file="RIF_nc model mean.svg" "RIF_nc model.contour"
celltool plot_contours --color-by=points --label-points --output-file="RIF_30min model mean.svg" "RIF_30min model.contour"
celltool plot_contours --color-by=points --label-points --output-file="RIF_60min model mean.svg" "RIF_60min model.contour"

#----measure areas
celltool measure_contours --output-file="RIF_nc model.csv" --area --centroid --shape-modes "RIF_nc model.contour" 1 2 - Aligned/*nc*.contour
celltool measure_contours --output-file="RIF_30min model.csv" --area --centroid --shape-modes "RIF_30min model.contour" 1 2 - Aligned/*30min*.contour
celltool measure_contours --output-file="RIF_60min model.csv" --area --centroid --shape-modes "RIF_60min model.contour" 1 2 - Aligned/*60min*.contour

#---plot area density plots
celltool plot_distribution --x-column=2 --output-file="time course area comparison.svg" "RIF_nc model.csv" "RIF_30min model.csv" "RIF_60min model.csv" Aligned/*.contour

# align contours, the RIF_nc model is used as reference in this scenario
celltool align_contours --reference="RIF_nc model.contour" --destination=Aligned Contours/*.contour
celltool plot_distribution --x-column=2 --y-column=3 --output-file="pca time course RIF_.svg" "RIF_nc model-normalized-positions.csv" "RIF_30min model-normalized-positions.csv" "RIF_60min model-normalized-positions.csv" Aligned/*.contour

#----measure curvature 
celltool measure_contours --output-file="RIF_nc curve.csv" --curvature - Aligned/*nc*.contour
celltool measure_contours --output-file="RIF_after 30min curve.csv" --curvature - Aligned/*30min*.contour
celltool measure_contours --output-file="RIF_after 60min curve.csv" --curvature - Aligned/*60min*.contour

#---plot curvature density plots
celltool plot_distribution --x-column=2 --output-file="time course RIF_curvature Positions.svg" "RIF_nc curve.csv" "RIF_after 30min curve.csv" "RIF_after 60min curve.csv" Aligned/*.contour
```
